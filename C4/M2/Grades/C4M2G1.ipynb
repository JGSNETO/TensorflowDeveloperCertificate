### Question 1
**What does MAE stand for?**

- Mean Average Error
- Mean Advanced Error
- Mean Absolute Error
- Mean Active Error

**Answer**: Mean Absolute Error  
**Explanation**: MAE (Mean Absolute Error) is a measure used to evaluate the accuracy of a model's predictions. It calculates the average of the absolute differences between the predicted values and the actual values.

### Question 2
**What’s the correct line of code to split an n column window into n-1 columns for features and 1 column for a label?**

- `dataset = dataset.map(lambda window: (window[n-1], window[1]))`
- `dataset = dataset.map(lambda window: (window[:-1], window[-1:]))`
- `dataset = dataset.map(lambda window: (window[-1:], window[:-1]))`

### Question 3
**What is a windowed dataset?**

- The time series aligned to a fixed shape
- There’s no such thing
- A fixed-size subset of a time series
- A consistent set of subsets of a time series

**Answer**: A fixed-size subset of a time series  
**Explanation**: A windowed dataset refers to a subset of a time series that is created by "sliding" a fixed-size window across the series. Each window is a smaller chunk of data from the original time series.

### Question 10
**What does ‘drop_remainder=True’ do?**

- It ensures that all data is used
- It ensures that all rows in the data window are the same length by adding data
- It ensures that the data is all the same shape
- It ensures that all rows in the data window are the same length by cropping data

**Answer**: It ensures that all rows in the data window are the same length by cropping data  
**Explanation**: When `drop_remainder=True`, any remaining data that doesn't fit the specified batch size will be dropped, ensuring that all batches have the same size.

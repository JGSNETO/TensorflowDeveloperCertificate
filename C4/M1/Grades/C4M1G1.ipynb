### Question 1
**What is a trend?**

- An overall consistent flat direction for data
- An overall consistent upward direction for data
- An overall consistent downward direction for data
- An overall direction for data regardless of direction

**Answer**: An overall direction for data regardless of direction  
Explanation: A trend refers to the long-term movement or direction of data over time, which could be upward, downward, or flat.

---

### Question 2
**In the context of time series, what is noise?**

- Unpredictable changes in time series data
- Sound waves forming a time series
- Data that doesn’t have a trend
- Data that doesn’t have seasonality

**Answer**: Unpredictable changes in time series data  
Explanation: Noise refers to random fluctuations or errors in the data that do not follow any specific pattern or trend.

---

### Question 3
**What is an example of a Univariate time series?**

- Fashion items
- Baseball scores
- Hour by hour weather
- Hour by hour temperature

**Answer**: Hour by hour temperature  
Explanation: A univariate time series involves observations of a single variable recorded over time at regular intervals.

---

### Question 4
**What is autocorrelation?**

- Data that follows a predictable shape, even if the scale is different
- Data that automatically lines up seasonally
- Data that doesn’t have noise
- Data that automatically lines up in trends

**Answer**: Data that follows a predictable shape, even if the scale is different  
Explanation: Autocorrelation refers to the correlation of a time series with a lagged version of itself, showing repeating patterns.

---

### Question 5
**What is a non-stationary time series?**

- One that moves seasonally
- One that is consistent across all seasons
- One that has a disruptive event breaking trend and seasonality
- One that has a constructive event forming trend and seasonality

**Answer**: One that moves seasonally  
Explanation: Non-stationary time series have statistical properties that change over time, such as seasonality or trends.

---

### Question 6
**What is imputed data?**

- A good prediction of future data
- A bad prediction of future data
- Data that has been withheld for various reasons
- A projection of unknown (usually past or missing) data

**Answer**: A projection of unknown (usually past or missing) data  
Explanation: Imputed data refers to estimates used to replace missing or incomplete values in a dataset.

---

### Question 7
**A sound wave is a good example of time series data**

- True
- False

**Answer**: True  
Explanation: A sound wave represents a sequence of values (amplitude) recorded at regular intervals, making it a time series.

---

### Question 8
**What is Seasonality?**

- Data aligning to the 4 seasons of the calendar
- A regular change in shape of the data
- Weather data
- Data that is only available at certain times of the year

**Answer**: A regular change in shape of the data  
Explanation: Seasonality refers to recurring patterns or cycles in data at regular intervals due to seasonal or periodic factors.
